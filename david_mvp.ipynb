```python
import pandas as pd
import numpy as np
from datetime import datetime
import pickle
```

```python
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('train.csv')
```

```python
from david_code import datetime_cleans
```

```python
df = datetime_cleans(df)
```

```python
df.head()
```

```python
data = df.drop(columns=['Title', 'Address.1st.Line', 'Address.2nd.Line', 'State', 'Zip.Code', 'License.Type', 'License.Issue.Date', 'Dental.School', 
                'Graduation.Date'])
```

```python
data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

```python
y = data['convert']
data = data.drop(columns=['convert', 'approach.date'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,y)
random_forest_model = RandomForestClassifier(n_estimators=300, max_depth=3, verbose=1)
random_forest_model.fit(X_train, y_train)
p_random_forest = random_forest_model.predict_proba(X_test)
random_forest_ll = log_loss(y_test, p_random_forest )
print("Random Forest Log Loss " + str(random_forest_ll))
```

```python
train_columns = data.columns
columns = set(data.columns)
```

```python
pickling_on = open("rfc.pickle","wb")
pickle.dump(random_forest_model, pickling_on)
pickling_on.close()
```
