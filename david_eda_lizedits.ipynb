```python
months = ['month_January', 'month_February', 'month_March', 'month_April', 'month_May', 'month_June',
              'month_July', 'month_August','month_September', 'month_October', 'month_November', 'month_December']
```

```python
def find_month_percents(df):
    result = []
    months = ['month_January', 'month_February', 'month_March', 'month_April', 'month_May', 'month_June',
              'month_July', 'month_August','month_September', 'month_October', 'month_November', 'month_December']
    for month in months:
        data = df[df[month] == 1]
        pct = sum(data.loc[:, 'convert'])/len(data)
        result.append((month, pct))
    return result
```

```python
month_pcts = find_month_percents(data)
```

```python
def find_day_percents(df):
    result = []
    days = ['day_of_week_Monday', 'day_of_week_Tuesday','day_of_week_Wednesday', 'day_of_week_Thursday', 
            'day_of_week_Friday','day_of_week_Saturday', 'day_of_week_Sunday']
    for day in days:
        data = df[df[day] == 1]
        pct = sum(data.loc[:, 'convert'])/len(data)
        result.append((day, pct))
    return result
```

```python
day_pcts = find_day_percents(data)
days = ['day_of_week_Monday', 'day_of_week_Tuesday','day_of_week_Wednesday', 'day_of_week_Thursday', 
            'day_of_week_Friday','day_of_week_Saturday', 'day_of_week_Sunday']
```

```python
y = data['convert']
data = data.drop(columns=['convert', 'approach.date'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,y)
```

```python
random_forest_model = RandomForestClassifier(n_estimators=300, max_depth=3, verbose=1)
```

```python
random_forest_model.fit(X_train, y_train)
p_random_forest = random_forest_model.predict_proba(X_test)
random_forest_ll = log_loss(y_test, p_random_forest )
print("Random Forest Log Loss " + str(random_forest_ll))
```
